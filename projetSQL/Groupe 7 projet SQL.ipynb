```python
import pandas as pd

# Liste tous les fichiers dans le répertoire courant qui se terminent par '.csv'
files = ['./airports.csv', './flights.csv']


dfAirports = pd.read_csv(files[0])

dfFlights = pd.read_csv(files[1])

dfAirports 
```

## Question 1

Combien y-a-t-il : 

- d’aéroports en tout, de départ et de destination
- Combien y-a-t-il d’aéroports où on ne passe pas à l’heure d’été (indice : colonne dst : 23) ? de fuseaux horaires (10 voir colone tzone dont une  « N »)
- de compagnies, d’avions, de vols annulés ? 


```python
# Nettoie les données en supprimant les doublons
dfAirports = dfAirports.drop_duplicates()

# compter le nombre d'aeroport
print('Nombre d\'aéroport :', dfAirports['name'].count())
```

```python
dfFlights
```

```python
# Nettoie les données en supprimant les doublons
dfFlights = dfFlights.drop_duplicates()

# compter le nombre d'aeroport
print('Nombre de depart :', dfFlights['dep_time'].count())

# nombre de d'arriver
print('Nombre d\'arriver :', dfFlights['arr_time'].count())

# Nombres d'aeroport où on ne passe pas à l'heure d'été (indice : colonne dst : 23)
print('Nombre d\'aeroport où on ne passe pas à l\'heure d\'été :', dfAirports[dfAirports['dst'] == 'N'].count()['dst'])

# Nombre de companies
print('Nombre de companies :', dfFlights['carrier'].nunique())

# Nombre d'avions
print('Nombre d\'avions :', dfFlights['tailnum'].nunique())

# Nombre de vols annulés

resultats = pd.DataFrame({
    'Nb de départs': [dfFlights['dep_time'].count()],
    'Nb d\'arrivées': [dfFlights['arr_time'].count()],
    'Nb d\'aéroports où on ne passe pas à l\'heure d\'été': [dfAirports[dfAirports['dst'] == 'N'].count()['dst']],
    'Nb de compagnies': [dfFlights['carrier'].nunique()],
    'Nb d\'avions': [dfFlights['tailnum'].nunique()],
    'Nb de vols annulés': ""
})

resultats
```

## Question 02
- Quel est l’aéroport de départ le plus emprunté ? 
- Quelles sont les 10 destinations les plus (moins) prisées (en indiquant le nom complet des destinations et non juste le code correspondant et le % corresp pour chaque destination) ?
- Quelle sont les 10 avions qui ont le plus (moins) décollé ?

```python
# L’aéroport de départ le plus emprunté ?
print ('L’aéroport de départ le plus emprunté :', dfFlights['origin'].value_counts().idxmax())

# Les 10 destinations les plus (moins) prisées (en indiquant le nom complet des destinations et non juste le code correspondant et le % corresp pour chaque destination), prend le dfAirports pour avoir le nom complet des destinations
print('Les 10 destinations les plus prisées :')

top_10_destinations = dfFlights['dest'].value_counts().head(10).index
top_10_destinations_df = dfAirports[dfAirports['faa'].isin(top_10_destinations)]
top_10_destinations_df = top_10_destinations_df.set_index('faa').loc[top_10_destinations]
top_10_destinations_df['percentage'] = dfFlights['dest'].value_counts(normalize=True).head(10).values * 100

print(top_10_destinations_df[['name', 'percentage']])

print ('\n\nLes 10 destinations les moins prisées :')

moin_10_destinations = dfFlights['dest'].value_counts().tail(10).index
moin_10_destinations_df = dfAirports[dfAirports['faa'].isin(moin_10_destinations)]
moin_10_destinations_df = moin_10_destinations_df.set_index('faa').loc[moin_10_destinations]
moin_10_destinations_df['percentage'] = dfFlights['dest'].value_counts(normalize=True).tail(10).values * 100

print(moin_10_destinations_df[['name', 'percentage']])
```

```python
filtered_dfFlights = dfFlights.dropna(subset=['dep_time'])

# Les 10 avions qui ont le plus (moins) décollé
print('Les 10 avions qui ont le plus décollé :\n')
print(filtered_dfFlights['tailnum'].value_counts().head(10))

print ('\n\nLes 10 avions qui ont le moins décollé :')
print(filtered_dfFlights['tailnum'].value_counts().tail(10))
```

## Question 3

- Combien chaque compagnie a desservi de destination ; 
- combien chaque compagnie a desservie de destination par aéroport d’origine. 
- Réaliser les graphiques adéquats qui synthétisent ces informations ? 

```python
# Nombre de compagnies desservant chaque dest
print('Nombre de compagnies desservant chaque dest :\n')
print(dfFlights.groupby('dest')['carrier'].nunique().sort_values(ascending=False))

# Nombre de destinations desservies par chaque compagnie par aéroport d'origine
print('Nombre de destinations desservies par chaque compagnie par aéroport d\'origine :\n')
print(dfFlights.groupby(['origin', 'carrier'])['dest'].nunique().sort_values(ascending=False))

# resultats = pd.DataFrame({
#     "Nb compagnie desserve chaque dest" : dfFlights.groupby('dest')['carrier'].nunique().sort_values(ascending=False),
#     "Nb "
# })
```
